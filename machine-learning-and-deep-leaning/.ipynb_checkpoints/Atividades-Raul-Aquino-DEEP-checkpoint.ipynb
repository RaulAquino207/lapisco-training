## Questão 1 - O que seria um problema não linearmente separável? Explique utilizando o problema da porta lógica XOR. Argumente sobre o motivo do Perceptron de Rosenblatt (1958) não conseguir resolver este tipo de problema.

`👆 Acima temos do lado esquerdo (AND) um exemplo de um problema linearmente separável e do lado direito (XOR) outro exemplo que representa um problema NÃO linearmente separável, ou seja, com que apenas 1 reta eu não consigo separar minhas classes em 2 áreas distintas que as definem. O perceptron de Rosenblatt só é capaz de resolver problemas linearmente separáveis pois só consegue traçar 1 reta.`

## Questão 2 - Por que uma das estratégias para treinamento da rede neural é baseado no gradiente descendente e qual a relação do gradiente com a atualização dos pesos dos neurônios.

![Url](https://miro.medium.com/max/600/0*_SzFcwNRNqFnyH7K.gif "gradiente")

`Nas redes neurais um dos nossos desafios é chegar no valor mínimo da nossa função custo (mínimo global), e para alcançarmos tal valor mínimo temos que fazer atualizações nos nossos vetores de pesos, que a acada iteração são atualizados tendo em vista essa minimização, ou seja, o gradiente que nos diz com qual valor atualizar os pesos. Buscando essa minimização temos um termo bem conhecido chamado OTIMIZAÇÃO, os otimizadores baseados no gradiente descedentes nos permitem convergir para o mínimo da função custo. Nos temos uma co-relação entre a taxa de aprendizagem e o gradiente nessa busca, onde fazendo uma comparação bem simples, nossa taxa de aprendizagem seria a aceleração da bolinhas azul do exemplo acima e o gradiente a direção para onde ela deve seguir.`

## Questão 3 - Considerando o dataset diabetes.csv, o mesmo utilizado nas atividades das aulas de machine learning, faça o treinamento e avalie as métricas de teste, para os modelos Shallow e Deep vistos em aula. Compare os resultados considerando as métricas: Acurácia, Tempo de treino e tempo de teste em relação aos classificadores MLP, KNN e QDA (vistos na aula de machine learning). Crie uma tabela com as métricas e liste os resultados, comente brevemente os resultados apresentados. Considere a Tabela 1 1 para listar os resultados.

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.layers import Flatten, Activation, Conv2D, MaxPool2D, Dense
from keras.optimizers import SGD, Adam
from tensorflow.keras import backend
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, f1_score
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
import time
```

```python
df_diabetes = pd.read_csv('diabetes.csv')
df_diabetes
```

```python
X = df_diabetes.iloc[:, 0:8]
y = df_diabetes.iloc[:, -1]
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

y = labelencoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train
```

- Shallow

```python
#Building shallow model: One Layer with sigmoid function and with 8 entries, which define the shape of my vector X.
shallowModel = Sequential()
shallowModel.add(Dense(1, input_shape=(8,), activation='sigmoid'))
shallowModel.compile(Adam(lr=0.05), 'binary_crossentropy', metrics=['accuracy'])
```

```python
shallow_training_start = time.time()
shallowModel.fit(X_train, y_train, epochs=200, verbose=1)
shallow_training_end = time.time()
shallow_training_total = shallow_training_end - shallow_training_start
print('shallow training : ', shallow_training_total)
```

```python
shallow_test_start = time.time()
results=shallowModel.evaluate(X_test, y_test)
shallow_test_end = time.time()
shallow_test_total = shallow_test_end - shallow_test_start
print('shallow test : ', shallow_test_total)
```

```python
print('Train ACC SHALLOW: ', results[1])
```

```python
metrics_shallow = []
metrics_shallow.extend((results[1], shallow_training_total, shallow_test_total))
metrics_shallow
```

- Deep

```python
#Building shallow model: With more than one hidden layer.
deepModel = Sequential()
deepModel.add(Dense(4, input_shape=(8,), activation='tanh')) 
deepModel.add(Dense(2, input_shape=(4,), activation='tanh'))
deepModel.add(Dense(1, input_shape=(2,), activation='sigmoid'))
deepModel.compile(optimizer=Adam(lr=0.05), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
deep_training_start = time.time()
deepModel.fit(X_train, y_train, epochs=200, verbose=1)
deep_training_end = time.time()
deep_training_total = deep_training_end - deep_training_start
print('deep training : ', deep_training_total)
```

```python
deep_test_start = time.time()
deepResults = deepModel.evaluate(X_test, y_test)
deep_test_end = time.time()
deep_test_total = deep_test_end - deep_test_start
print('deep test : ', deep_test_total)
```

```python
print('Train ACC DEEP: ', deepResults[1])
```

```python
metrics_deep = []
metrics_deep.extend((deepResults[1], deep_training_total, deep_test_total))
metrics_deep
```

- MLP

```python
#Building MLP model: With optimization method(ADAM), and adaptive learning rate.
clf_mlp = MLPClassifier(solver='adam', learning_rate='adaptive', hidden_layer_sizes = (100,), 
                        max_iter=1300, learning_rate_init=5e-04, tol=1e-4, random_state = 1)

mlp_training_start = time.time()
clf_mlp.fit(X_train, y_train)
mlp_training_end = time.time()
mlp_training_total = mlp_training_end - mlp_training_start
print('mlp training : ', mlp_training_total)

mlp_test_start = time.time()
y_predicted_mlp = clf_mlp.predict(X_test)
mlp_test_end = time.time()
mlp_test_total = mlp_test_end - mlp_test_start
print('mlp test : ', mlp_test_total)


acc_mlp = accuracy_score(y_test, y_predicted_mlp)
print('Train ACC MLP: ', acc_mlp)
```

```python
metrics_mlp = []
metrics_mlp.extend((acc_mlp, mlp_training_total, mlp_test_total))
metrics_mlp
```

- KNN

```python
#Building KNN model: With the number of neighbors equal to 3.
neigh = KNeighborsClassifier(n_neighbors=3)

knn_training_start = time.time()
neigh.fit(X_train, y_train)
knn_training_end = time.time()
knn_training_total = knn_training_end - knn_training_start
print('knn training : ', knn_training_total)

knn_test_start = time.time()
y_predicted_knn = neigh.predict(X_test)
knn_test_end = time.time()
knn_test_total = knn_test_end - knn_test_start
print('knn test : ', knn_test_total)

acc_knn = accuracy_score(y_test, y_predicted_knn)
print('Train ACC KNN: ', acc_knn)
```

```python
metrics_knn = []
metrics_knn.extend((acc_knn, knn_training_total, knn_test_total))
metrics_knn
```

- QDA

```python
#Building QDA model: A statistical classifier
clf_qda = QuadraticDiscriminantAnalysis(store_covariance=True)

qda_training_start = time.time()
clf_qda.fit(X_train, y_train)
qda_training_end = time.time()
qda_training_total = qda_training_end - qda_training_start
print('qda training : ', qda_training_total)

qda_test_start = time.time()
y_predicted_qda = clf_qda.predict(X_test)
qda_test_end = time.time()
qda_test_total = qda_test_end - qda_test_start
print('qda test : ', qda_test_total)

acc_qda = accuracy_score(y_test, y_predicted_qda)
print('Train ACC QDA: ', acc_qda)
```

```python
metrics_qda = []
metrics_qda.extend((acc_qda, qda_training_total, qda_test_total))
metrics_qda
```

```python
data_q3 = []
data_q3.append(metrics_shallow)
data_q3.append(metrics_deep)
data_q3.append(metrics_mlp)
data_q3.append(metrics_knn)
data_q3.append(metrics_qda)
```

```python
columns = ['Acurácia', 'Tempo de treino', 'Tempo de teste']
columns_id_q3 = ['Shallow', 'Deep', 'MLP', 'KNN', 'QDA']
metrics_matrix_q3 = pd.DataFrame(data=data_q3, columns=columns)
metrics_matrix_q3.set_index(x for x in columns_id_q3)
```

## 👆 Resultados 📝
`Como é facilmente observado os métodos mais simples de classificação, como KNN, QDA e MLP conseguiram um nível de acurácia bem parelho aos modelos que utilizam camadas densas, podemos usar até mesmo o tempo de treino como uma métrica muito importante à se analisar, com menos de 1 segundo de tempo de treino os métodos mais simples tiveram até mais acurácia que os modelos que gastaram 7 segundos.`

## Questão 4 - Considerando o dataset MNIST (importe a partir do keras.datasets), dataset dos dígitos manuscritos utilizado nos exemplos de CNN em aula. Faça o treinamento e avalie as métricas de treino e teste, para os modelos Fully Connected Model e Convolutional Neural Network, vistos em aula e compare os resultados considerando as métricas: Acurácia, Tempo de treino e tempo de teste em relação aos classificadores MLP, KNN e QDA (para estes classificadores, lembre-se de utilizar a vetorização da imagem). Crie uma tabela com as métricas e liste os resultados, comente brevemente os resultados apresentados. Considere a Tabela 1 para listar os resultados.

```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
from keras.utils.np_utils import to_categorical

(X_train, y_train), (X_test, y_test) = mnist.load_data(path='/tmp/mnist.npz')
print(X_train.shape)
print(X_test.shape)
```

```python
plt.imshow(X_train[150], cmap='gray')
plt.show()
```

- Fully Connected Model

```python
X_train = X_train.reshape(-1, 28*28)
X_test = X_test.reshape(-1, 28*28)

X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255

print(X_train.shape)
print(X_test.shape)
```

```python
y_train_categ = to_categorical(y_train) 
y_test_categ = to_categorical(y_test)

print(y_train[100])
print(y_train_categ[100])
```

```python
model_fully = Sequential()
model_fully.add(Dense(512, input_dim=28*28, activation='relu'))
model_fully.add(Dense(256, activation='relu'))
model_fully.add(Dense(128, activation='relu'))
model_fully.add(Dense(32, activation='relu'))
model_fully.add(Dense(10, activation='softmax'))

model_fully.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
fully_training_start = time.time()
history_fully = model_fully.fit(X_train, y_train_categ, batch_size=128, epochs=10, verbose=1, validation_split=0.3)
fully_training_end = time.time()
fully_training_total = fully_training_end - fully_training_start
```

```python
fully_test_start = time.time()
results_fully = model_fully.evaluate(X_test, y_test_categ)
fully_test_end = time.time()
fully_test_total = fully_test_end - fully_test_start
```

```python
metrics_fully = []
metrics_fully.extend((results_fully[1], fully_training_total, fully_test_total))
metrics_fully
```

```python
plt.plot(history_fully.history['accuracy'])
plt.plot(history_fully.history['val_accuracy'])
plt.legend(['Training', 'Validation'])
plt.title('Accuracy')
plt.xlabel('Epochs')
```

- CNN

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data(path='/tmp/mnist.npz')

X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)
```

```python
backend.clear_session()

#AM: Feature extraction stages
model_cnn = Sequential()
model_cnn.add(Conv2D(64, (3,3), input_shape=(28, 28, 1), padding=('valid')))
model_cnn.add(MaxPool2D(pool_size=(2,2)))
model_cnn.add(Activation('relu'))

model_cnn.add(Flatten())

#AM: Fully connected stages (Classification)
model_cnn.add(Dense(128, activation='relu'))
model_cnn.add(Dense(10, activation='softmax'))

model_cnn.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model_cnn.summary()
```

```python
cnn_train_start = time.time()
history_CNN = model_cnn.fit(X_train, y_train_categ, batch_size=128, epochs=10, verbose=1, validation_split=0.3)
cnn_train_end = time.time()
cnn_train_total = cnn_train_end - cnn_train_start
```

```python
cnn_test_start = time.time()
results_cnn = model_cnn.evaluate(X_test, y_test_categ)
cnn_test_end = time.time()
cnn_test_total = cnn_test_end - cnn_test_start
```

```python
metrics_cnn = []
metrics_cnn.extend((results_cnn[1], cnn_train_total, cnn_test_total))
metrics_cnn
```

```python
plt.plot(history_CNN.history['accuracy'])
plt.plot(history_CNN.history['val_accuracy'])
plt.legend(['Training', 'Validation'])
plt.title('Accuracy')
plt.xlabel('Epochs')
```

```python
data_q4 = []
data_q4.append(metrics_fully)
data_q4.append(metrics_cnn)
data_q4.append(metrics_mlp)
data_q4.append(metrics_knn)
data_q4.append(metrics_qda)
```

```python
columns = ['Acurácia', 'Tempo de treino', 'Tempo de teste']
columns_id_q4 = ['Fully', 'CNN', 'MLP', 'KNN', 'QDA']
metrics_matrix_q4 = pd.DataFrame(data=data_q4, columns=columns)
metrics_matrix_q4.set_index(x for x in columns_id_q4)
```

## 👆 Resultados 📝
`Agora vemos uma grande diferença dos modelos que usam o esquema sequential dos mais simples, eles tiveram uma acurácia alta , cerca de 98%, porém com um tempo de treinamento bem maior, porém esse tempo compensa pela taxa de acurácia bem mais acertiva que os métodos mais simples`
