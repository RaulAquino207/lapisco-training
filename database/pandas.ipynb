```python
df_iris.iloc[:, 0:4].describe()
```

```python
values = df_iris.iloc[:, 0:4]
std_all_dataset = []
    
for i in values:
    for j in values[i]:
        std_all_dataset.append(j)
        
std_of_dataset = np.std(std_all_dataset) * 3
print('Desvio padrão em relação a toda base {}'.format(std_of_dataset))

arr_values = df_iris.iloc[:, 0:4].values.tolist()
arr_class = df_iris.iloc[:, 4].values.tolist()
print('before', len(arr_values), len(arr_class))
for i, v in enumerate(arr_values):
    pop = False
#     print(v)
    for j in v:
#         print(j)
        if j > std_of_dataset:
            pop = True
    if pop == True:
        del(arr_values[i])
        del(arr_class[i])
        pop = False
#         print('VALUES ABOVE DELETED')
        
print('after', len(arr_values), len(arr_class))
```

2) Faça um merge de dois dataframes e crie uma nova coluna para indicar a que dataframe pertence cada linha. ✅

#### ⬇️ Para facilitar dividi um mesmo dataframe 2 e depois os juntei.

```python
half = len(df_titanic) // 2
# Dividindo o dataset em 2 partes
df_titanic_part_1 = df_titanic.iloc[:half, :]
df_titanic_part_1 = df_titanic_part_1.assign(Part = [1 for x in range(0,len(df_titanic_part_1))])
df_titanic_part_2 = df_titanic.iloc[half:, :]
df_titanic_part_2 = df_titanic_part_2.assign(Part = [2 for x in range(0,len(df_titanic_part_2))])

df_titanic_full = pd.concat((df_titanic_part_1, df_titanic_part_2), axis=0)
df_titanic_full
```

4) Transforme os dados contínuos de idade do titanic em dados categóricos de acordo com a seguinte regra: ✅

0 a 18 anos → Criança

18 a 65 anos → Adulto

maior de 65 → Idoso


```python
df_titanic_full_values = df_titanic_full.values.tolist()
for i, v in enumerate(df_titanic_full_values):
    if v[4] <= 18:
        df_titanic_full_values[i][4] = 'Criança'
    elif v[4] > 18 and v[4] <= 65:
        df_titanic_full_values[i][4] = 'Adulto'
    elif v[4] > 65:
        df_titanic_full_values[i][4] = 'Idoso'

df = pd.DataFrame(df_titanic_full_values,columns=['PassengerId','Survived','Name', 'Sex', 'Age', 'Part'])
df
```

```python
df.Sex = df.Sex.astype('category')
df.Age = df.Age.astype('category')
df.info()
```

5) Aplique o hold out utilizando diferentes porcentagens de divisão para separar a base iris em treino e teste (e.g. 60/40, 70/30, 80/20, 90/10). Crie vetores para armazenar os dados da divisão de treino e teste da seguinte forma: ✅

X_train → armazena os dados de treino

X_test → armazena os dados de teste

y_train → armazena a label correspondente dos dados de treino

y_test → armazena a label correspondente dos dados de teste

```python
# Usando sklearn.model_selection
from sklearn.model_selection import train_test_split
iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X, y, test_size=0.4)
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X, y, test_size=0.3)
X_train_3, X_test_3, y_train_3, y_test_3 = train_test_split(X, y, test_size=0.2)
X_train_4, X_test_4, y_train_4, y_test_4 = train_test_split(X, y, test_size=0.1)
```

```python
# Sem usar o sklearn.model_selection
def hold_out(X, y, test_size=0.3):
    X_train = []
    X_test = []
    y_train = []
    y_test = []
    tam = len(X) * (1 - test_size)
    for idx, value in enumerate(X):
        if idx <= tam:
            aux = []
            for bla in value:
                aux.append(bla)
            X_train.append(aux)
            y_train.append(y[idx])
        else:
            X_test.append(value)
            y_test.append(y[idx])
        
        
    return X_train, X_test, y_train, y_test

    
    
X_train, X_test, y_train, y_test = hold_out(X, y)
```
