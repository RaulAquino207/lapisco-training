```python
def remove_parenteses(item):
    if '(' in item:
        return item.replace('(', '').replace(')', '')
    else:
        item

names = df_titanic.Name.apply(remove_parenteses)
for idx, value in enumerate(names):
    if value != None:
        df_titanic['Name'][idx] = value
```

1) Verifique a presença de outliers (e.g., valor maior que 3 vezes o desvio padrão da base) na base de dados iris ou outra base de dados. Caso não exista, insira uma ou mais linhas e selecione elas utilizando a regra mencionada. ✅

```python
df_iris.iloc[:, 0:4].describe()
```

```python
values = df_iris.iloc[:, 0:4]
std_all_dataset = []
    
for i in values:
    for j in values[i]:
        std_all_dataset.append(j)
        
std_of_dataset = np.std(std_all_dataset) * 3
print(std_of_dataset)

arr_values = df_iris.iloc[:, 0:4].values.tolist()
arr_class = df_iris.iloc[:, 4].values.tolist()
print('before', len(arr_values))
for i, v in enumerate(arr_values):
    pop = False
    print(v)
    for j in v:
        print(j)
        if j > std_of_dataset:
            pop = True
    if pop == True:
        del(arr_values[i])
        del(arr_class[i])
        pop = False
        print('VALUES ABOVE DELETED')
        
print('after', len(arr_values), len(arr_class))
```
